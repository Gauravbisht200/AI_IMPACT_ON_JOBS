### INTRODUCING A VARIABLE COMMULATIVE_AI_AUTOMATION
project's core challenge arises from the presence of two critical columns in the dataset: "AI Impact" and "AI Workload." These two columns presented a significant dilemma due to the lack of a discernible deciding factor that would facilitate prioritization.so we Introduce a new column COMMULATIVE_AI_AUTOMATION -now the more the value of COMMULATIVE_AI_AUTOMATION the more it is benificial for the company to Integrate AI with that particular job.

## DATA OF Communication & PR DOMAIN

```python
CP=grouped_data.get_group('Communication & PR')
pd.set_option('display.max_rows', 10)
CP['COMMULATIVE_AI_AUTOMATION']=CP['AI Impact'] * CP['AI_Workload_Ratio']
columns_to_print =['Job titiles','AI Impact','AI_Workload_Ratio','COMMULATIVE_AI_AUTOMATION']
CP[columns_to_print].sort_values(by='COMMULATIVE_AI_AUTOMATION')
```
